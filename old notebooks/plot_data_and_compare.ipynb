```python
import dipoles
from tqdm import tqdm
import frenel
import force

eps_Si = frenel.get_interpolate('Si')
eps_Au = frenel.get_interpolate('Au')

STOP = 45
R = 60
dist = 20
point = [0,0,dist+R]
angle = 25*np.pi/180
phase = 0
a = 0.5
wl=800


F_theory =  force.F(wl, 1, point, R, eps_Si, angle, 1, phase, a, STOP)

print(F_theory)
```
